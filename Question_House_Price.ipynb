### Q3: If sales price is greater/less than 2 standard deviation from the mean sale price then delete such rows?
* How the shape of data will change after removing these values ?

```python
q1=data['Sale_Price'].quantile(.25)
q1
```

```python

q3=data['Sale_Price'].quantile(.75)
q3
```

```python
iqr=q3-q1
iqr
```

```python
upperlimit=q3+1.5*iqr
lowerlimit=q1-1.5*iqr
(lowerlimit,upperlimit)
```

```python
count=0
def limit_imputer(value):
    if value>upperlimit :
        count=count+1
        return np.nan
    if value<lowerlimit :
        count=count+1
        return np.nan
    else :
        return value
```

```python
data['Sale_Price']=data['Sale_Price'].apply(limit_imputer)
```

```python
print("Before deleting outliers length = " , (len(data['Sale_Price'])))                
target_mean = data['Sale_Price'].mean()
target_mean 
```

```python
target_sd = data['Sale_Price'].std()
target_sd 
```

```python
print("After deleting outliers length = " ,  ((len(data['Sale_Price']))-count))
data 
```
